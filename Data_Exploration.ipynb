```python
# def find_overall_sentiment_trump(row):
#     if row.absa_neutral > .95:
#         return "neutral"
#     elif row.absa_positive > row.absa_negative:
#         return "pro_trump"
#     else:
#         return "anti_trump"
        
        

# def find_overall_sentiment_biden(row):
#     if row.absa_neutral > .95:
#         return "neutral"
#     elif row.absa_positive > row.absa_negative:
#         return "pro_biden"
#     else:
#         return "anti_biden"
    
def find_overall_sentiment_trump(row):
    if row.vader_neutral > .95:
        return "neutral"
    elif row.vader_pos > row.vader_negative:
        return "pro_trump"
    else:
        return "anti_trump"
        
        

def find_overall_sentiment_biden(row):
    if row.vader_neutral > .95:
        return "neutral"
    elif row.vader_pos > row.vader_negative:
        return "pro_biden"
    else:
        return "anti_biden"
        
    
    
    
    

trump["sentiment"] = trump.apply(find_overall_sentiment_trump, axis=1)
biden["sentiment"] = biden.apply(find_overall_sentiment_biden, axis=1)
```
